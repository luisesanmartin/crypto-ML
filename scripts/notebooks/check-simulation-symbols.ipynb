```python
folder = '../../data/results/'
```

```python
prefix = 'simulations-manual-rules-symbols-step'
files = [f'{folder}{file}' for file in os.listdir(folder)if prefix in file]
```

```python
dfs = [pd.read_csv(file) for file in files]
```

```python
df = pd.concat(dfs)
```

```python
df = df.sort_values('Profits', ascending=False).reset_index(drop=True)
```
